<h2>Algorytm drzewa dycyzyjnego<h2>



Algorytm drzewa decyzyjnego jest jednym z najpopularniejszych algorytmów uczenia maszynowego. 
Do rozwiązania konkretnego problemu wykorzystuje strukturę przypominającą drzewo oraz ich możliwe kombinacje. 
Należy on do klasy algorytmów uczenia nadzorowanego.

Drzewo decyzyjne to struktura zawierająca węzeł główny, gałęzie i węzły liściowe. Każdy węzeł wewnętrzny oznacza test atrybutu, każda gałąź oznacza wynik testu, a każdy węzeł liści zawiera etykietę klasy. Najwyżej położonym węzłem w drzewie jest węzeł główny.

Podczas implementacji algorytmu drzewa decyzyjnego przyjęto pewne założenia. Są one wymienione poniżej

Na początku cały zbiór treningowy jest traktowany jako korzeń.
Wartości cech muszą być kategoryczne. Jeżeli wartości są ciągłe, to są dyskretyzowane przed budową modelu.
Rekordy są rozdzielane rekursywnie na podstawie wartości atrybutów.
Kolejność umieszczania atrybutów jako korzeni lub węzłów wewnętrznych drzewa jest dokonywana przy użyciu pewnego podejścia statystycznego.


<h2>Czy drzewo decyzyjne mozna zastosowac do klasyfikacji?<h2>


Algorytm drzew decyzyjnych jest obecnie znany pod swoją nowoczesną nazwą CART, która oznacza Classification and Regression Trees (Drzewa klasyfikacji i regresji). Classification and Regression Trees lub CART to termin wprowadzony przez Leo Breimana w odniesieniu do algorytmów drzew decyzyjnych, które można stosować do rozwiązywania problemów klasyfikacji i modelowania regresji.

Algorytm CART stanowi podstawę dla innych ważnych algorytmów, takich jak workowe drzewa decyzyjne, las losowy i wzmocnione drzewa decyzyjne. W tym jądrze będę rozwiązywał problem klasyfikacji. Dlatego też algorytm ten będę określał również jako problem klasyfikacji drzew decyzyjnych.

<h2>Terminologia drzewa decyzyjnego<h2>



W algorytmie drzewa decyzyjnego istnieje struktura przypominająca drzewo, w której każdy węzeł wewnętrzny reprezentuje test atrybutu, każda gałąź reprezentuje wynik testu, a każdy węzeł liścia reprezentuje etykietę klasy. Ścieżki od węzła głównego do węzła liścia reprezentują reguły klasyfikacji.

Można zauważyć, że algorytm drzewa decyzyjnego wiąże się z pewną terminologią. Pojęcia związane z algorytmem drzewa decyzyjnego są następujące

<h3>Podstawowe węzły<h3>

*Węzeł główny(Root Node)
Reprezentuje całą populację lub próbę. Następnie zostaje ona podzielona na dwa lub więcej jednorodnych zbiorów.<br>

*Węzeł decyzyjny(decision node)
Gdy węzeł podrzędny dzieli się na kolejne węzły podrzędne, nazywany jest węzłem decyzyjnym.<br>

*Węzeł liściowy/końcowy(leaf/terminal node)
Węzły, które nie ulegają podziałowi, są nazywane węzłami liściowymi lub końcowymi.

<h3>Inne potrzebne terminy<h3>

*Podział(Splitting)
Jest to proces dzielenia węzła na dwa lub więcej węzłów podrzędnych.<br>

*Przycinanie(pruning)
Proces usuwania węzłów podrzędnych węzła decyzyjnego nazywamy przycinaniem. Jest to proces odwrotny do splittingu.

*Gałąź/poddrzewo(Branch/subtree)
Pododcinek całego drzewa nazywamy gałęzią lub poddrzewem.<br>

*Węzeł nadrzędny i węzeł podrzędny(Parent and child node)
Węzeł, który jest podzielony na węzły podrzędne, jest nazywany węzłem nadrzędnym węzłów podrzędnych, gdzie węzły podrzędne są dziećmi węzła nadrzędnego.


<h2>Działanie drzewa decyzyjnego<h2>

Dla każdego atrybutu w zbiorze danych algorytm drzewa decyzyjnego tworzy węzeł. Najważniejszy atrybut jest umieszczany w węźle głównym.

W celu oceny danego zadania zaczynamy od węzła głównego i schodzimy w dół drzewa, podążając za węzłem, który spełnia nasz warunek lub decyzję.

Proces ten jest kontynuowany aż do osiągnięcia węzła liścia. Zawiera on prognozę lub wynik drzewa decyzyjnego.

<h2>Wybór atrybutów-ważne<h2>

Podstawowym wyzwaniem w implementacji drzewa decyzyjnego jest określenie atrybutów, które uznajemy za węzeł główny i każdy kolejny poziom. Proces ten nazywany jest selekcją atrybutów. Istnieją różne miary selekcji atrybutów służące do identyfikacji atrybutu, który może być uznany za węzeł główny na każdym poziomie.

Istnieją dwie popularne miary selekcji atrybutów. Są one następujące: 

*Information gain-Stosując to jako kryterium, zakładamy, że atrybuty są kategoryczne

*Gini Index-natomiast w przypadku wskaźnika Giniego atrybuty są ciągłe



<h2>Information gain<h2>

Stosując "informatio gain" jako kryterium, próbujemy oszacować informacje, zawarte w każdym z atrybutów. Aby zrozumieć pojęcie "information gain", musimy poznać inne pojęcie, zwane entropią.



<h3>Entropia<h3>


Entropia mierzy nieczystość w danym zbiorze danych. W fizyce i matematyce entropia jest określana jako losowość lub niepewność zmiennej losowej X. W teorii informacji odnosi się ona do nieczystości w grupie przykładów. Przyrost informacji to spadek entropii. Przyrost informacji oblicza różnicę między entropią przed podziałem a średnią entropią po podziale zbioru danych na podstawie podanych wartości atrybutów.

Entropia jest przedstawiana za pomocą następującego wzoru:

Tutaj c jest liczbą klas, a pi jest prawdopodobieństwem związanym z i-tą klasą.

Algorytm drzewa decyzyjnego ID3 (Iterative Dichotomiser) wykorzystuje entropię do obliczania przyrostu informacji. Obliczając spadek miary entropii dla każdego atrybutu, można więc obliczyć przyrost informacji. Atrybut o największym przyroście informacji jest wybierany jako atrybut rozdzielający w węźle.

<h3>gini index<h3>

Inną miarą selekcji atrybutów  jest indeks Giniego. Do tworzenia punktów podziału wykorzystuje on metodę Giniego.

Wskaźnik Giniego można przedstawić za pomocą następującego diagramu:

Również w tym przypadku c jest liczbą klas, a pi jest prawdopodobieństwem związanym z i-tą klasą.

Wskaźnik Giniego mówi, że jeśli wybierzemy losowo dwa elementy z populacji, to muszą one należeć do tej samej klasy, a prawdopodobieństwo tego wynosi 1, jeśli populacja jest czysta.

Działa z kategoryczną zmienną docelową "Sukces" lub "Porażka". Wykonuje tylko podziały binarne. Im wyższa wartość współczynnika Giniego, tym większa homogeniczność. CART (Classification and Regression Tree) wykorzystuje metodę Giniego do tworzenia podziałów binarnych.


W przypadku atrybutu o wartości dyskretnej jako atrybut podziału wybierany jest podzbiór, który daje minimalny wskaźnik Giniego dla tego wybranego atrybutu. W przypadku atrybutów o ciągłej wartości strategia polega na wybraniu każdej pary sąsiadujących wartości jako możliwego punktu podziału, a punkt o mniejszym indeksie giniego wybiera się jako punkt podziału. Atrybut o minimalnym indeksie Giniego jest wybierany jako atrybut dzielący.

```python

Post-Pruning
W postpruningu zagłębiamy się coraz głębiej w drzewo, aby zbudować kompletne drzewo. Jeżeli drzewo wykazuje problem nadmiernego dopasowania, wówczas w kroku postpruningowym wykonuje się przycinanie. Do sprawdzenia efektu przycinania używamy danych z walidacji krzyżowej. Używając danych walidacji krzyżowej, sprawdzamy, czy rozszerzenie węzła spowoduje poprawę, czy nie. Jeśli poprawa jest widoczna, możemy kontynuować rozbudowę tego węzła. Jeśli natomiast węzeł ten wykaże spadek dokładności, nie należy go rozbudowywać. Należy więc przekształcić ten węzeł w węzeł liściasty.
```

```python
# encode variables with ordinal encoding

encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])


X_train = encoder.fit_transform(X_train)

X_test = encoder.transform(X_test)
```

<h2> Decision Tree Classifier with criterion gini index <h2>

```python
# import DecisionTreeClassifier

from sklearn.tree import DecisionTreeClassifier
```

```python
# instantiate the DecisionTreeClassifier model with criterion gini index

clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=0)


# fit the model
clf_gini.fit(X_train, y_train)
```

<h3>Przewidywanie wyników zestawu testowego za pomocą kryterium współczynnika gini.<h3>

```python
y_pred_gini = clf_gini.predict(X_test)
```

<h3>Sprawdź wynik dokładności za pomocą kryterium wskaźnika giniego<h3>

```python
from sklearn.metrics import accuracy_score

print('Model accuracy score with criterion gini index: {0:0.4f}'. format(accuracy_score(y_test, y_pred_gini)))
```

W tym przypadku y_test to prawdziwe etykiety klas, a y_pred_gini to przewidywane etykiety klas w zbiorze testowym.

<h3>Porównanie dokładności zestawu treningowego i testowego<h3>

```python
y_pred_train_gini = clf_gini.predict(X_train)

y_pred_train_gini
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train_gini)))
```

<h3>Check for overfitting and underfitting<h3>

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(clf_gini.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(clf_gini.score(X_test, y_test)))
```

W tym przypadku wynik dokładności zestawu treningowego wynosi 0,7865, a dokładności zestawu testowego - 0,8021. Te dwie wartości są dość porównywalne. Nie ma więc oznak niedopasowania.

<h2>Macierz konfuzji<h2>

Macierz konfuzji jest narzędziem służącym do podsumowania wydajności algorytmu klasyfikacji. Macierz konfuzji daje nam jasny obraz wydajności modelu klasyfikacyjnego oraz rodzajów błędów popełnianych przez model. Daje ona podsumowanie poprawnych i niepoprawnych przewidywań w podziale na poszczególne kategorie. Podsumowanie jest przedstawiane w formie tabelarycznej.

Podczas oceny wydajności modelu klasyfikacyjnego możliwe są cztery rodzaje wyników. Te cztery wyniki są opisane poniżej

Prawdziwe wyniki pozytywne (TP) - prawdziwe wyniki pozytywne występują wtedy, gdy przewidujemy, że obserwacja należy do pewnej klasy i faktycznie należy ona do tej klasy.

Prawdziwe negatywy (TN) - Prawdziwe negatywy występują, gdy przewidujemy, że obserwacja nie należy do pewnej klasy, a obserwacja faktycznie do niej nie należy.

Fałszywe wyniki pozytywne (FP) - Fałszywe wyniki pozytywne występują, gdy przewidujemy, że obserwacja należy do pewnej klasy, ale w rzeczywistości nie należy do tej klasy. Ten rodzaj błędu nazywany jest błędem typu I.

Fałszywe negatywy (FN) - fałszywe negatywy występują wtedy, gdy przewidujemy, że obserwacja nie należy do pewnej klasy, ale w rzeczywistości należy do tej klasy. Jest to bardzo poważny błąd i nazywany jest błędem II rodzaju.

Te cztery wyniki są podsumowane w poniższej macierzy konfuzji.

```python
# Print the Confusion Matrix and slice it into four pieces

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_en)

print('Confusion matrix\n\n', cm)
```
